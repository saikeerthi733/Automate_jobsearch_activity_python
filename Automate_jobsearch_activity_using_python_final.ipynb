```python
import requests
from bs4 import BeautifulSoup
sheet=[]
r=requests.get("https://www.monster.com/jobs/search/?q=Automation-Engineer&where=UnitedStates")
soup = BeautifulSoup(r.content, 'html.parser')
jobs=soup.find(id='ResultsContainer')
job_list=jobs.find_all('section',class_='card-content')
for item in job_list:
    d={}
    try:
        d['job_company']=item.find('div',class_='company').text.strip('\n')
    except:
        d['job_company']=None
        continue
    try:
        d['job_location']=item.find('div',class_='location').text.strip('\r\n')
    except:
        d['job_location']=None
        continue
    try:
        d['job_posted_on']=item.find('time').text
    except:
        d['job_posted_on']=None
        continue

    sheet.append(d)
print(sheet)
```
